```python
#we have 3types to iterate over rows: iterrows(), iteritems(),itertuples().
#iterrows: returns each index value along with a series containing the data in each row.
#iteritems: this function iterates over each column as key, value pair with the label as key, and column value as a Series object.
#itertuples:it returns tuple. The first element of the tuple will be the row’s corresponding index value, while the remaining values are the row values.

for i, j in df2.iterrows():
    print(i, j)
    print()
```
